```python
import tweepy
from tweepy import OAuthHandler
 
# Twittter App Credentials
consumer_key = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
consumer_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
access_token = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
access_token_secret = "XXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXXX"
 
# Calling API
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth)
 



    
```

# RANDOM TWEETS

```python
# Here keywords for extracting random tweets has been used
import pandas as pd
keyword1 = ["great","life","feeling","win","live","best","cool","hope","know","lol","music","school","college","fun","today","want","time","back","call","look","first","please","everyone","amazing","summer","rain","winter","wait","cause","coming","even","odd","maybe","head","big","small","soon","later","run","lot","show","least","most","yeah","way","hi","hello"]
 
# Fetching tweets using twitter API cons
# maximum count would be 100 default is 15
## No tweets can be extracted older than a week 
## historical tweets can't be extracted
tweets1=[]
for i in keyword1:
    tweets1.append(tweepy.Cursor(api.search,q=i, lang='en', exclude='retweets',tweet_mode="extended", until='2020-08-18', result_type='mixed',geocode="21.146,79.088,1650km").items())
    
id1=[]
datetime=[]
text=[]
username=[]
userloc=[]

for i in range(len(tweets1)):
    for info in tweets1[i]:
        id1.append(info.id)
        datetime.append(info.created_at)
        text.append(info.full_text)
        username.append(info.user.screen_name)
        userloc.append(info.user.location)
        
    
    
        

df2=pd.DataFrame()
df2['tweet_ID']=id1
df2['user_Name']=username
df2['timeline']=datetime
df2['location']=userloc
df2['tweet_Text']=text


    
df2
    
```
